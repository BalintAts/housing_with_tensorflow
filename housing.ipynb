### Preprocessing fro Model

```python
X = df.drop('price',axis = 1).values   #vlues: we need numppy array, not dataframe
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)  #we don't fit to the test set, becouse the model should haveknowledge about 
# only the train set
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
X_train.shape
```

```python
df.columns
```

```python
model.add(Dense(19,activation="relu"))
model.add(Dense(19,activation="relu"))
model.add(Dense(19,activation="relu"))
model.add(Dense(19,activation="relu"))

model.add(Dense(1))

model.compile(optimizer='adam', loss= 'mse')
```

```python
model.fit(x=X_train, y=y_train, 
          validation_data=(X_test, y_test),
         batch_size=128,epochs = 400)  #validation happens after eachepoch, and logs out the error, so we can see, but it not counts into the training
# smaller batches avoids overfitting, but the training is slower
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
plt.figure(figsize=[15,15])
losses.plot()
```

```python
# if val data begins to rise, it means the model is overfitting on the training data
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
mean_squared_error(y_test,predictions)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
mean_absolute_error(y_test,predictions)
```

```python
explained_variance_score(y_test,predictions)
```

```python
plt.scatter(y_test,predictions)
plt.plot(y_test,y_test)
```

```python
# outliers ruins the model
mean_absolute_error(y_test,predictions)/predictions.mean()
```
